```python
#Function that will help me with calculating the distance between two points in a trajectory
#using Haversine distance calculated in meters(m)
def haversine_m(lat1, lon1, lat2, lon2):
    R = 6371000.0
    phi1, phi2 = map(radians, (lat1, lat2))
    dphi = radians(lat2 - lat1)
    dlambda = radians(lon2 - lon1)
    a = sin(dphi/2)**2 + cos(phi1)*cos(phi2)*sin(dlambda/2)**2
    return 2 * R * asin(min(1, sqrt(a)))
```

```python
#Function that returns a dataframe with imputed distance, time difference, horizontal speed and vertical rate in every row (between every consecutive point)
def pointwise(traj):
    #Return DataFrame with ts,lat,lon,alt_FL,dt,dist_m,speed_m_s,vertical_rate_FL_s
    df = pd.DataFrame(traj, columns=['t','lat','lon','alt_FL']).astype(float)
    df['dt'] = df['t'].diff().fillna(1.0)
    # distances
    dists = [0.0]
    for i in range(1, len(df)):
        dists.append(haversine_m(df.loc[i-1,'lat'], df.loc[i-1,'lon'],
                                 df.loc[i,'lat'], df.loc[i,'lon']))
    df['dist_m'] = dists
    # speeds
    df['speed_m_s'] = df['dist_m'] / df['dt'].replace(0, np.nan)
    df['speed_m_s'] = df['speed_m_s'].fillna(0.0)
    # vertical rate
    df['vertical_rate_FL_s'] = df['alt_FL'].diff() / df['dt'].replace(0, np.nan)
    df['vertical_rate_FL_s'] = df['vertical_rate_FL_s'].fillna(0.0)
    return df
```

```python
#Function that returns a trajectory with calculated features that I think are important to help me build a model for detecting bad trajectories.

def summarize(traj):
    df = pointwise(traj)
    feats = {}
    feats['mean_speed_m_s'] = float(df['speed_m_s'].mean())
    feats['std_speed_m_s'] = float(df['speed_m_s'].std())
    feats['max_speed_m_s'] = float(df['speed_m_s'].max())
    feats['mean_vrate'] = float(df['vertical_rate_FL_s'].mean())
    feats['max_abs_vrate'] = float(np.abs(df['vertical_rate_FL_s']).max())
    feats['total_dist_m'] = float(df['dist_m'].sum())
    feats['sinuosity'] = feats['total_dist_m'] / haversine_m(df.loc[0,'lat'], df.loc[0,'lon'], df.loc[len(df)-1,'lat'], df.loc[len(df)-1,'lon'])
    feats['frac_zero_speed'] = float((df['speed_m_s'] == 0).mean())
    feats['n_large_alt_jumps'] = int((df['alt_FL'].diff().abs() > 50).sum())
    return feats
```

```python
#Analyzes a trajectory and explains potential problems based on learned model features
def explain_trajectory(traj, model, thresholds=None):
    feats = summarize(traj)

    feats = {k: v for k, v in feats.items() if k not in ['frac_zero_speed', 'n_large_alt_jumps']}

    X_sample = pd.DataFrame([feats])

    pred = model.predict(X_sample)[0]
    print("Trajectory Analysis")
    print("-------------------------")
    print(f"Predicted label: {'GOOD' if pred==1 else 'BAD'}")

    issues = []
    if feats['max_speed_m_s'] > 754:
        issues.append("Extremely high max speed (possible GPS spike)")
    if feats['std_speed_m_s'] > 165:
        issues.append("Large speed variation (unstable trajectory)")
    if feats['sinuosity'] > 2.89:
        issues.append("Path is too curved or inconsistent")
    if abs(feats['mean_vrate']) > 15:
        issues.append("Excessive vertical rate (possible altitude jumps)")
    if feats['total_dist_m'] < 1664:
        issues.append("Very short movement distance (possible incomplete track)")
    if feats['total_dist_m'] > 26000:
        issues.append("Very long movement distance")

    if issues:
        print("Potential Issues:")
        for i in issues:
            print(" -", i)
    else:
        print("Trajectory appears smooth and consistent.")
```

Starting to build percentage estimation model here.

```python
def bearing_deg(lat1, lon1, lat2, lon2):
    #Forward azimuth A->B in degrees [0, 360)
    1, 2 = radians(lat1), radians(lat2)
    d = radians(lon2 - lon1)
    x = sin(d) * cos(2)
    y = cos(1) * sin(2) - sin(1) * cos(2) * cos(d)
    return (degrees(atan2(x, y)) + 360.0) % 360.0

def angle_diff_deg(a, b):
    #Smallest signed angular difference a->b in [-180, 180]
    return (b - a + 180.0) % 360.0 - 180.0
```

```python
def pointwise_dir(traj):
    df = pd.DataFrame(traj, columns=['t', 'lat', 'lon', 'alt_FL']).astype(float)

    # time delta
    df['dt'] = df['t'].diff().fillna(1.0)
    df.loc[df['dt'] == 0, 'dt'] = 1.0  # guard

    dists = [0.0]
    for i in range(1, len(df)):
        dists.append(haversine_m(df.loc[i - 1, 'lat'], df.loc[i - 1, 'lon'],
                                 df.loc[i, 'lat'], df.loc[i, 'lon']))
    df['dist_m'] = dists
    df['speed_m_s'] = df['dist_m'] / df['dt']

    # vertical rate (FL/s)
    df['vertical_rate_FL_s'] = df['alt_FL'].diff() / df['dt']
    df['vertical_rate_FL_s'] = df['vertical_rate_FL_s'].fillna(0.0)

    # bearing A->B
    brngs = [np.nan]
    for i in range(1, len(df)):
        brngs.append(bearing_deg(df.loc[i - 1, 'lat'], df.loc[i - 1, 'lon'],
                                 df.loc[i, 'lat'], df.loc[i, 'lon']))
    df['bearing_deg'] = brngs

    # turn and turn-rate
    turns = [np.nan]
    for i in range(1, len(df)):
        a = df.loc[i - 1, 'bearing_deg']
        b = df.loc[i, 'bearing_deg']
        turns.append(np.nan if (np.isnan(a) or np.isnan(b)) else angle_diff_deg(a, b))
    df['turn_deg'] = turns
    df['turn_rate_deg_s'] = df['turn_deg'] / df['dt']

    return df
```

```python
#Dropping distance feature and adding direction feature
def summarize_dir_full(traj):
    df = pointwise_dir(traj)

    feats = {}
    feats['mean_speed_m_s'] = float(df['speed_m_s'].mean())
    feats['std_speed_m_s'] = float(df['speed_m_s'].std())
    feats['max_speed_m_s'] = float(df['speed_m_s'].max())

    feats['mean_vrate'] = float(df['vertical_rate_FL_s'].mean())
    feats['max_abs_vrate'] = float(np.abs(df['vertical_rate_FL_s']).max())

    total_dist = float(df['dist_m'].sum())  # internal calculation only
    lat1, lon1 = df.loc[0, 'lat'], df.loc[0, 'lon']
    lat2, lon2 = df.loc[len(df) - 1, 'lat'], df.loc[len(df) - 1, 'lon']
    crow = haversine_m(lat1, lon1, lat2, lon2)
    feats['sinuosity'] = float(total_dist / crow) if crow > 0 else 1.0

    turn = df['turn_deg'].dropna()
    trate = df['turn_rate_deg_s'].dropna()

    feats['mean_abs_turn_deg'] = float(np.abs(turn).mean() if len(turn) else 0.0)
    feats['max_abs_turn_deg'] = float(np.abs(turn).max() if len(turn) else 0.0)
    feats['mean_abs_turn_rate_deg_s'] = float(np.abs(trate).mean() if len(trate) else 0.0)

    return feats
```

```python
features_list_dir, labels_dir = [], []

for traj in good_trajectories:
    features_list_dir.append(summarize_dir_full(traj))
    labels_dir.append(1)

for traj in bad_trajectories:
    features_list_dir.append(summarize_dir_full(traj))
    labels_dir.append(0)

df_features_dir = pd.DataFrame(features_list_dir)
df_features_dir['label'] = labels_dir
df_features_dir = df_features_dir.sample(frac=1, random_state=42).reset_index(drop=True)
display(df_features_dir.head())
```

```python
good_only = df_features_dir[df_features_dir['label'] == 1].drop(columns='label')
good_only = good_only.replace([np.inf, -np.inf], np.nan).dropna()

mu_good = good_only.mean(numeric_only=True)
std_good = good_only.std(numeric_only=True).replace(0, 1e-9)

quality_features = [c for c in df_features_dir.columns if c != 'label']
```

```python
def quality_from_features(feats):
    penalties = []

    # z-score distance from good (beyond 2 std)
    for feat in quality_features:
        x = feats[feat]
        mu = mu_good[feat]
        sd = std_good[feat]

        z = abs(x - mu) / sd
        excess = max(0, z - 2)   # anything beyond 2 std is penalized
        penalty = min(excess / 3, 1)  # saturate by ~5 std
        penalties.append(penalty)

    avg_penalty = np.mean(penalties)
    quality = (1 - avg_penalty) * 100
    return round(float(np.clip(quality, 0, 100)), 2)
```

```python
def get_quality_percent(traj):
    feats = summarize_dir_full(traj)
    return quality_from_features(feats)
```

```python
# Universal "normal" ranges for the direction-based features
UNIVERSAL_NORMAL_RANGES = {
    # speed statistics (m/s)
    'mean_speed_m_s': (80.0, 280.0),  # typical cruise-ish range
    'std_speed_m_s': (0.0, 80.0),  # too high => very unstable speeds
    'max_speed_m_s': (100.0, 320.0),  # >320 m/s ~ >1150 km/h => suspicious

    # vertical motion (FL/s)
    'mean_vrate': (-10.0, 10.0),  # average climb/descent rate
    'max_abs_vrate': (0.0, 25.0),  # extremely high vertical rate is odd

    # geometry / turning
    'sinuosity': (1.0, 2.5),  # >2.5 means very wiggly / backtracking
    'mean_abs_turn_deg': (0.0, 25.0),  # average turn angle per step
    'max_abs_turn_deg': (0.0, 120.0),  # single extreme U-turns
    'mean_abs_turn_rate_deg_s': (0.0, 4.0),  # how fast we turn per second
}


def quality_from_features_universal(feats):

    penalties = []

    for feat, (low, high) in UNIVERSAL_NORMAL_RANGES.items():
        x = feats[feat]

        # within normal range  no penalty
        if low <= x <= high:
            penalties.append(0.0)
            continue

        # how far outside the range?
        if x < low:
            diff = low - x
            span = max(low, 1e-6)  # scale by lower bound
        else:  # x > high
            diff = x - high
            span = max(high - low, 1e-6)  # scale by range width

        rel = diff / span
        penalty = min(rel, 1.0)  # cap at 1
        penalties.append(penalty)

    if not penalties:
        return 100.0

    avg_penalty = float(np.mean(penalties))
    quality = (1.0 - avg_penalty) * 100.0
    return round(float(np.clip(quality, 0.0, 100.0)), 2)


def get_quality_percent_aggressive(traj):

    feats = summarize_dir_full(traj)
    return quality_from_features_universal(feats)
```

Optimization of the trajectory data.

```python
#Functions for removing point based on exceeding limits

def remove_velocity_outliers(df, max_speed):

   #Remove points where the segment speed exceeds max_speed (m/s).

    if len(df) < 2:
        return df

    high_speed_mask = [False]  # first point always kept

    for i in range(1, len(df)):
        dist = haversine_m(
            df.loc[i - 1, 'lat'], df.loc[i - 1, 'lon'],
            df.loc[i, 'lat'], df.loc[i, 'lon']
        )
        time_diff = df.loc[i, 'timestamp'] - df.loc[i - 1, 'timestamp']
        speed = dist / time_diff if time_diff > 0 else 0.0
        high_speed_mask.append(speed > max_speed)

    high_speed_mask = np.array(high_speed_mask)
    return df[~high_speed_mask].reset_index(drop=True)


def optimize_sinuosity(df, max_sinuosity):
    #Remove points that cause strong backtracking to reduce sinuosity.

    if len(df) < 4:
        return df

    # Direct (crow-flight) distance startend
    direct_dist = haversine_m(
        df.iloc[0]['lat'], df.iloc[0]['lon'],
        df.iloc[-1]['lat'], df.iloc[-1]['lon']
    )
    if direct_dist == 0:
        # All points at the same location; nothing to optimize
        return df

    # Current path length
    total_dist = 0.0
    for i in range(1, len(df)):
        total_dist += haversine_m(
            df.iloc[i - 1]['lat'], df.iloc[i - 1]['lon'],
            df.iloc[i]['lat'], df.iloc[i]['lon']
        )

    current_sinuosity = total_dist / direct_dist if direct_dist > 0 else 1.0
    if current_sinuosity <= max_sinuosity:
        return df

    # Angle-based pruning of backtracking points
    keep = [True] * len(df)
    keep[0] = True
    keep[-1] = True

    for i in range(1, len(df) - 1):
        v1_lat = df.iloc[i]['lat'] - df.iloc[i - 1]['lat']
        v1_lon = df.iloc[i]['lon'] - df.iloc[i - 1]['lon']
        v2_lat = df.iloc[i + 1]['lat'] - df.iloc[i]['lat']
        v2_lon = df.iloc[i + 1]['lon'] - df.iloc[i]['lon']

        mag1 = np.sqrt(v1_lat ** 2 + v1_lon ** 2)
        mag2 = np.sqrt(v2_lat ** 2 + v2_lon ** 2)

        if mag1 == 0 or mag2 == 0:
            # Zero vector  ignore (keep point)
            continue

        dot = v1_lat * v2_lat + v1_lon * v2_lon
        cos_angle = dot / (mag1 * mag2)
        cos_angle = np.clip(cos_angle, -1.0, 1.0)
        angle = np.arccos(cos_angle)  # radians

        # >135 = backtracking / sharp U-turn
        if angle > np.radians(135):
            keep[i] = False

    df_opt = df[keep].reset_index(drop=True)
    return df_opt
```

Two optimization modes: aggressive + stats-based

```python
#Agressive
def optimize_trajectory_aggressive(trajectory):

    #Aggressive optimization with fixed thresholds:
     # - Speed limit: 300 m/s
     # - Sinuosity limit: 1.5

    df = pd.DataFrame(trajectory, columns=['timestamp', 'lat', 'lon', 'alt_FL'])

    # Sort + deduplicate by timestamp
    df = df.sort_values('timestamp').drop_duplicates(subset=['timestamp']).reset_index(drop=True)

    # Remove crazy GPS spikes (very high speed)
    df = remove_velocity_outliers(df, max_speed=300.0)

    # Reduce path backtracking / curvature
    if len(df) > 8:
        df = optimize_sinuosity(df, max_sinuosity=1.5)

    return df.values
```

```python
#Statistcs extraction
def calculate_real_statistics(good_trajectories, bad_trajectories):

   # Compute average feature statistics for good and bad trajectories

    good_feats = []
    for traj in good_trajectories:
        try:
            good_feats.append(summarize(traj))
        except Exception:
            continue

    bad_feats = []
    for traj in bad_trajectories:
        try:
            bad_feats.append(summarize(traj))
        except Exception:
            continue

    df_good = pd.DataFrame(good_feats)
    df_bad = pd.DataFrame(bad_feats)

    print(f"Analyzed {len(df_good)} good and {len(df_bad)} bad trajectories")

    real_good_stats = df_good.mean().to_dict()
    real_bad_stats = df_bad.mean().to_dict()

    print("Good trajectories statistics:")
    for k, v in real_good_stats.items():
        print(f"  {k}: {v:.2f}")

    print("\nBad trajectories statistics:")
    for k, v in real_bad_stats.items():
        print(f"  {k}: {v:.2f}")

    return real_good_stats, real_bad_stats
```

```python
def optimize_trajectory_stats_based(trajectory, good_stats):

    #Optimization using thresholds derived from GOOD trajectories:
     # - Speed limit: mean_good_max_speed * 1.1
     # - Sinuosity limit: mean_good_sinuosity * 1.2

    df = pd.DataFrame(trajectory, columns=['timestamp', 'lat', 'lon', 'alt_FL'])

    # Sort + deduplicate by timestamp
    df = df.sort_values('timestamp').drop_duplicates(subset=['timestamp']).reset_index(drop=True)

    # Speed limit from good trajectories (with small buffer)
    max_speed = good_stats['max_speed_m_s'] * 1.1
    df = remove_velocity_outliers(df, max_speed=max_speed)

    # Sinuosity limit from good trajectories (with small buffer)
    if len(df) > 8:
        max_sinuosity = good_stats['sinuosity'] * 1.2
        df = optimize_sinuosity(df, max_sinuosity=max_sinuosity)

    return df.values
```

Creating functions for evaluation of both models for optimization

```python
def traj_to_model_output(traj, model):
    """
    From a single trajectory:
      - compute features (using summarize)
      - drop unused ones
      - get model prediction + probabilities
    Returns: (features_dict, pred_label, proba_array)
    """
    feats = summarize(traj)
    feats_filtered = {
        k: v for k, v in feats.items()
        if k not in ['frac_zero_speed', 'n_large_alt_jumps']
    }
    X = pd.DataFrame([feats_filtered])
    pred = model.predict(X)[0]
    proba = model.predict_proba(X)[0]
    return feats, pred, proba
```

```python
#Function for comparing unoptimized and optimized trajectory
def compare_trajectories_detailed(original, optimized,
                                  original_label="Original",
                                  optimized_label="Optimized"):
    df_orig = pd.DataFrame(original, columns=['timestamp', 'lat', 'lon', 'alt_FL']).sort_values('timestamp')
    df_opt = pd.DataFrame(optimized, columns=['timestamp', 'lat', 'lon', 'alt_FL']).sort_values('timestamp')

    def compute_speeds(df):
        speeds = [0.0]
        for i in range(1, len(df)):
            dist = haversine_m(
                df.iloc[i - 1]['lat'], df.iloc[i - 1]['lon'],
                df.iloc[i]['lat'], df.iloc[i]['lon']
            )
            dt = df.iloc[i]['timestamp'] - df.iloc[i - 1]['timestamp']
            speeds.append(dist / dt if dt > 0 else 0.0)
        return speeds

    orig_speeds = compute_speeds(df_orig)
    opt_speeds = compute_speeds(df_opt)

    feats_orig = summarize(original)
    feats_opt = summarize(optimized)

    comparison_features = ['mean_speed_m_s', 'max_speed_m_s', 'std_speed_m_s', 'sinuosity']
    orig_vals = [feats_orig[f] for f in comparison_features]
    opt_vals = [feats_opt[f] for f in comparison_features]

    fig, axes = plt.subplots(2, 3, figsize=(18, 10))

    # Altitude vs time
    axes[0, 0].plot(df_orig['timestamp'], df_orig['alt_FL'], 'ro-', alpha=0.7, label=original_label)
    axes[0, 0].plot(df_opt['timestamp'], df_opt['alt_FL'], 'go-', alpha=0.7, label=optimized_label)
    axes[0, 0].set_title("Altitude vs Time")
    axes[0, 0].set_xlabel("Time")
    axes[0, 0].set_ylabel("Altitude (FL)")
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Lat vs Lon
    axes[0, 1].plot(df_orig['lon'], df_orig['lat'], 'ro-', alpha=0.7, label=original_label)
    axes[0, 1].plot(df_opt['lon'], df_opt['lat'], 'go-', alpha=0.7, label=optimized_label)
    axes[0, 1].set_title("2D Trajectory (Lat-Lon)")
    axes[0, 1].set_xlabel("Longitude")
    axes[0, 1].set_ylabel("Latitude")
    axes[0, 1].legend()
    axes[0, 1].grid(True)

    # Speed per point
    axes[0, 2].plot(range(len(orig_speeds)), orig_speeds, 'ro-', alpha=0.7, label=original_label)
    axes[0, 2].plot(range(len(opt_speeds)), opt_speeds, 'go-', alpha=0.7, label=optimized_label)
    axes[0, 2].axhline(y=300, linestyle='--', alpha=0.5, label='300 m/s')
    axes[0, 2].set_title("Speed per Point")
    axes[0, 2].set_xlabel("Index")
    axes[0, 2].set_ylabel("Speed (m/s)")
    axes[0, 2].legend()
    axes[0, 2].grid(True)

    # Speed histogram
    axes[1, 0].hist(orig_speeds, bins=10, alpha=0.7, label=original_label)
    axes[1, 0].hist(opt_speeds, bins=10, alpha=0.7, label=optimized_label)
    axes[1, 0].axvline(x=300, linestyle='--', alpha=0.5, label='300 m/s')
    axes[1, 0].set_title("Speed Distribution")
    axes[1, 0].set_xlabel("Speed (m/s)")
    axes[1, 0].set_ylabel("Count")
    axes[1, 0].legend()
    axes[1, 0].grid(True)

    # Time differences
    orig_dt = df_orig['timestamp'].diff().dropna()
    opt_dt = df_opt['timestamp'].diff().dropna()

    axes[1, 1].plot(range(len(orig_dt)), orig_dt, 'ro-', alpha=0.7, label=original_label)
    axes[1, 1].plot(range(len(opt_dt)), opt_dt, 'go-', alpha=0.7, label=optimized_label)
    axes[1, 1].set_title("Time Intervals")
    axes[1, 1].set_xlabel("Segment index")
    axes[1, 1].set_ylabel("t (s)")
    axes[1, 1].legend()
    axes[1, 1].grid(True)

    # Feature comparison
    x = np.arange(len(comparison_features))
    width = 0.35
    axes[1, 2].bar(x - width / 2, orig_vals, width, alpha=0.7, label=original_label)
    axes[1, 2].bar(x + width / 2, opt_vals, width, alpha=0.7, label=optimized_label)
    axes[1, 2].set_xticks(x)
    axes[1, 2].set_xticklabels(comparison_features, rotation=45)
    axes[1, 2].set_title("Feature Comparison")
    axes[1, 2].set_ylabel("Value")
    axes[1, 2].legend()
    axes[1, 2].grid(True)

    plt.tight_layout()
    plt.show()

    print("=== COMPARISON STATISTICS ===")
    print(f"Original trajectory:  {len(original)} points")
    print(f"Optimized trajectory: {len(optimized)} points")
    print(f"Removed:              {len(original) - len(optimized)} points")
    print(f"Max speed (orig):     {max(orig_speeds):.2f} m/s")
    print(f"Max speed (opt):      {max(opt_speeds):.2f} m/s")
```

```python
#Evaluation of an optimizer
def evaluate_optimizer_on_all(trajectories, model, optimizer_fn, sample_size=None):

    if sample_size is None:
        sample_size = len(trajectories)

    results = []
    successful = 0

    print(f"Evaluation of {sample_size} trajectories...")
    print("=" * 60)

    for i in range(sample_size):
        original = trajectories[i]

        feats_orig, pred_orig, proba_orig = traj_to_model_output(original, model)

        try:
            optimized = optimizer_fn(original)
            feats_opt, pred_opt, proba_opt = traj_to_model_output(optimized, model)

            success = (pred_orig == 0 and pred_opt == 1)
            if success:
                successful += 1

            results.append({
                'traj_index': i,
                'original_points': len(original),
                'optimized_points': len(optimized),
                'points_removed': len(original) - len(optimized),
                'original_pred': pred_orig,
                'optimized_pred': pred_opt,
                'original_proba_bad': proba_orig[0],
                'original_proba_good': proba_orig[1],
                'optimized_proba_bad': proba_opt[0],
                'optimized_proba_good': proba_opt[1],
                'success': success,
                'original_max_speed': feats_orig['max_speed_m_s'],
                'optimized_max_speed': feats_opt['max_speed_m_s'],
                'original_total_dist': feats_orig['total_dist_m'],
                'optimized_total_dist': feats_opt['total_dist_m'],
            })

            if i % 50 == 0:
                print(f"Processed {i}/{sample_size}...")

        except Exception as e:
            # If something goes wrong with this trajectory, skip it
            # print(f"Error with trajectory {i}: {e}")
            continue

    results_df = pd.DataFrame(results)

    print("\n" + "=" * 60)
    print("OPTIMIZATION RESULTS")
    print("=" * 60)

    total_tested = len(results_df)
    success_rate = 100.0 * successful / total_tested if total_tested > 0 else 0.0

    print(f"Total tested trajectories: {total_tested}")
    print(f"Successfully optimized (BADGOOD): {successful}")
    print(f"Success rate: {success_rate:.2f}%")

    print("\nAverage number of points per trajectory:")
    print(f"  Original:  {results_df['original_points'].mean():.1f}")
    print(f"  Optimized: {results_df['optimized_points'].mean():.1f}")
    print(f"  Removed:   {results_df['points_removed'].mean():.1f}")

    stayed_bad = len(results_df[(results_df['original_pred'] == 0) & (results_df['optimized_pred'] == 0)])
    became_good = len(results_df[(results_df['original_pred'] == 0) & (results_df['optimized_pred'] == 1)])
    stayed_good = len(results_df[(results_df['original_pred'] == 1) & (results_df['optimized_pred'] == 1)])
    became_bad = len(results_df[(results_df['original_pred'] == 1) & (results_df['optimized_pred'] == 0)])

    print("\nDETAILED OVERVIEW:")
    print(f"  BAD  BAD: {stayed_bad}")
    print(f"  BAD  GOOD: {became_good}")
    print(f"  GOOD  GOOD: {stayed_good}")
    print(f"  GOOD  BAD: {became_bad}")

    return results_df
```

```python
#Function for optimization vizualization
def plot_optimization_results(results_df):
    fig, axes = plt.subplots(1, 2, figsize=(15, 6))

    # Pie chart: success vs failure
    success_count = results_df['success'].sum()
    fail_count = len(results_df) - success_count
    axes[0].pie(
        [success_count, fail_count],
        labels=['Successfully optimized', 'Not optimized'],
        autopct='%1.1f%%'
    )
    axes[0].set_title("Optimization Success Rate")

    # Scatter: original vs optimized GOOD probability
    axes[1].scatter(
        results_df['original_proba_good'],
        results_df['optimized_proba_good'],
        alpha=0.6,
        c=results_df['success'].map({True: 'green', False: 'red'})
    )
    axes[1].plot([0, 1], [0, 1], 'k--', alpha=0.3)
    axes[1].set_xlabel("P(GOOD) original")
    axes[1].set_ylabel("P(GOOD) optimized")
    axes[1].set_title("Change in Probability for GOOD Class")
    axes[1].grid(True)

    plt.tight_layout()
    plt.show()
```
